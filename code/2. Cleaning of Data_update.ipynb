The standard deviation from the mean is not large, which is a good sign that most data is consistent.  
However, there are some values which are too small, for example: left ankle, right ankle and left knee, and many of the min values.  
These will have to be cleaned.

```python
# creating copy so as not to corrupt original df
df1 = df_in1.copy()
```

```python
# checking again
df1.info()
```

```python
# checking again
df1.describe()
```

```python
# plotting of data to visualize histogram patterns
for column in df1.columns:
    plt.figure()
    sns.distplot(df1[column])
```

From the charts above, the following features need to be cleaned:

right knee  
left knee  
right shoulder  
left shoulder

## Removal of outliers

```python
# removing values below 3 x standard deviation below mean
df1_new = df1[df1["rightknee"] > (df1["rightknee"].mean() - 3*np.std(df1["rightknee"]))]
```

```python
# checking df
df1_new
```

```python
# check for new spread for right knee
df1_new['rightknee'].hist()
```

```python
# check for left knee as it was next for outlier removal, 
# but it seems the right knee removal has already removed the outliers for left knee
df1_new['leftknee'].hist()
```

```python
# check for right shoulder spread
df1_new['rightshoulder'].hist()
```

```python
# removal of values larger than 3 x std deviation above the mean
df1_new = df1_new[(df1_new['rightshoulder'] < df1_new['rightshoulder'].mean() + 3*np.std(df1_new['rightshoulder']))]
```

```python
# check df
df1_new
```

```python
# visualizing spread after outlier removal
df1_new['rightshoulder'].hist()
```

```python
# checking for left shoulder spread
df1_new['leftshoulder'].hist()
```

```python
# removal of values above mean + 3 x std deviation
df1_new = df1_new[(df1_new['leftshoulder'] < df1_new['leftshoulder'].mean() + 3*np.std(df1_new['leftshoulder']))]
```

```python
# check new spread plot
df1_new['leftshoulder'].hist()
```

```python
# final check on df
df1_new
```
