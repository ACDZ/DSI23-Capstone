## Trying out XGBoost model

```python
X = df_in.drop(labels = ['curve'], axis = 1)
y = df_in['curve']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
```

```python
D_train = xgb.DMatrix(X_train, label=Y_train)
D_test = xgb.DMatrix(X_test, label=Y_test)
```

```python
param = {
    'eta': 0.3, 
    'max_depth': 3,  
    'objective': 'multi:softprob',  
    'num_class': 3} 

steps = 20  # The number of training iterations
```

```python
model = xgb.train(param, D_train, steps)
```

```python
preds = model.predict(D_test)
best_preds = np.asarray([np.argmax(line) for line in preds])

print("Precision = {}".format(precision_score(Y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(Y_test, best_preds, average='macro')))
print("Accuracy = {}".format(accuracy_score(Y_test, best_preds)))
```

## XGBoost with Gridsearch CV

```python
clf = xgb.XGBClassifier()
parameters = {
     "eta"    : [0.05, 0.10, 0.15, 0.20, 0.25, 0.30 ] ,
     "max_depth"        : [ 3, 4, 5, 6, 8, 10, 12, 15],
     "min_child_weight" : [ 1, 3, 5, 7 ],
     "gamma"            : [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
     "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
     }

grid = GridSearchCV(clf,
                    parameters, n_jobs=4,
                    scoring="neg_log_loss",
                    cv=3)

grid.fit(X_train, Y_train)
```

```python
print(grid.best_score_)
```

#### Looks like gridsearch does not work very well with this dataset and model combination

The blended model provided superior accuracy, precision and recall as compared to the normal decision tree, bagged decision tree and boosted decision tree.
